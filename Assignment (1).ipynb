Q3.Explain the use of the following operators using an example:
(i) / (Division):
This operator performs regular division, dividing one number by another and returning the quotient (result of the division).

Example: 10 / 3 = 3.3333 (May vary based on programming language settings)

(ii) % (Modulo):
This operator finds the remainder after dividing one number by another. It returns the value that is leftover after the division is complete.

Example: 10 % 3 = 1 (10 divided by 3 leaves a remainder of 1)

(iii) // (Floor Division):
This operator performs integer division, dividing one number by another and returning the largest whole number less than or equal to the quotient. It discards the fractional part.

Example: 10 // 3 = 3 (10 divided by 3 is 3.3333, but floor division truncates to 3)

(iv) ** (Exponentiation):
This operator raises the left operand (base) to the power of the right operand (exponent). It calculates the base multiplied by itself the number of times specified by the exponent.

Example: 2 ** 3 = 8 (2 raised to the power of 3 is 2 x 2 x 2 = 8)

Q4. Using a while loop, verify if the number A is purely divisible by number B and if so then how many
times it can be divisible.

```python
a=int(input())
b=int(input())
count=0
while a%b==0:
    a//=b
    count+=1
if count>0:
    print(f"{a} is purely divisible by {b} and divisible {count} times")
else:
    print(f"{a} is not purely divisible by {b}.")
    
```
