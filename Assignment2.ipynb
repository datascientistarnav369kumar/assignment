3.How do you convert one data type to another in Python?
Ans:-we can covert one data type to another using built-in functions or constructors specific to each data type.
Example:-int():- Converts a value to an integer.
float():- Converts a value to a floating-point number.
str():- Converts a value to a string.
list():- Converts a sequence (e.g., tuple, string) to a list.
tuple():- Converts a sequence (e.g., list, string) to a tuple.
set():- Converts a sequence (e.g., list, tuple) to a set.
dict():- Converts a sequence of key-value pairs (e.g., list of tuples) to a dictionary.
